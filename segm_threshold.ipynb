```python
import numpy as np
import matplotlib.pyplot as plt
```

**[Task 1.2 (a).]()** Use `plt.imread` to load an image:

```python
img = plt.imread('data/NIH3T3/im/dna-0.png')
```

**[Task 1.2 (b).]()** Show the image:

```python
plt.imshow(img)
plt.colorbar()
```

**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
img2=np.zeros_like(img)
for i,j in np.ndindex(img.shape[0], img.shape[1]): 
    if img[i,j] > 0.1:
        img2[i,j] = 1
    else:
        img2[i,j] = 0

plt.imshow(img2)
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
img_gt=plt.imread('data/NIH3T3/gt/0.png')
```

Compute and print the Dice coefficient:

```python

dc=(2*(np.sum(img2*img_gt)))/(np.sum(img2)+np.sum(img_gt))
print(dc)
   
```
