**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def sobel_h(img):
    img_sobelh=np.zeros_like(img)
    maskeh =[(1,0,-1), (2,0,-2), (1, 0, -1)] 
    size=list((-1, 0, 1))
    b=list((0,1,2))
    d=list((0,1,2))
    for i,j in np.ndindex((img.shape[0]-2),(img.shape[1]-2)):
        matrix= np.zeros((3,3))
        #pixel=np.zeros((3,3))
        for x,a in zip(size, b):
            for y,c in zip(size,d):
                matrix[a,c]=img[i+1+x,j+1+y]
        pixel=((matrix*maskeh).sum())/8
        #print(pixel)
        img_sobelh[i,j]=pixel
    return img_sobelh






def sobel_v(img):
    img_sobelv=np.zeros_like(img)
    maskeh =[(1,2,1), (0,0,0), (-1, -2, -1)] 
    size=list((-1, 0, 1))
    b=list((0,1,2))
    d=list((0,1,2))
    for i,j in np.ndindex((img.shape[0]-2),(img.shape[1]-2)):
        matrix= np.zeros((3,3))
        #pixel=np.zeros((3,3))
        for x,a in zip(size, b):
            for y,c in zip(size,d):
                matrix[a,c]=img[i+1+x,j+1+y]
        pixel=((matrix*maskeh).sum())/8
        #print(pixel)
        img_sobelv[i,j]=pixel
    return img_sobelv
```

**[Task 2.3.]()** Test your implementations from above:

```python
img_sobelh=sobel_h(img)
img_sobelv=sobel_v(img_sobelh)
plt.imshow(img_sobelv)
plt.colorbar()
```
